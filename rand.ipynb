```python
import math

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


plt.style.use('seaborn')
%matplotlib inline
```

```python
SAMPLE_SIZE = 1000

u1 = np.random.uniform(0, 1, SAMPLE_SIZE)
u2 = np.random.uniform(0, 1, SAMPLE_SIZE)

u1_t = 2 * u1 - 1
u2_t = 2 * u2 - 1

m = u1_t **2 + u2_t **2
idx = np.where(m <= 1)
u1_t = u1_t[idx]
u2_t = u2_t[idx]
r = m[idx]

# uniform -> normal
y1 = u1_t * ((-2 * np.log(r**2)) / r ** 2) ** (0.5)
y2 = u2_t * ((-2 * np.log(r**2)) / r ** 2) ** (0.5)

# 
y1 = y1[y1 <= 5]
y1 = y1[y1 >= -5]
y2 = y2[y2 <= 5]
y2 = y2[y2 >= -5]
```

```python
fig1 = plt.figure(figsize=(10, 8))

ax1 = fig1.add_subplot(2, 1, 1)
ax1.hist(u1, color='red')
plt.title('Uniform Distribution')
plt.xlabel('u1')
plt.ylabel('Freq.')

ax2 = fig1.add_subplot(2, 1, 2)
ax2.hist(u2, color='green')
plt.title('Uniform Distribution')
plt.xlabel('u2')
plt.ylabel('Freq.')

plt.tight_layout()

fig2 = plt.figure(figsize=(10, 8))

ax1 = fig2.add_subplot(2, 1, 1)
ax1.hist(y1, color='red')
plt.title("Normal Distribution")
plt.xlabel("y1. sampled form u1")
plt.ylabel("Freq.")
ax2 = fig2.add_subplot(2, 1, 2)
ax2.hist(y2, color='green')
plt.title('Noraml Distribution')
plt.xlabel("y2. sampled from u2")
plt.ylabel("Freq.")

plt.tight_layout()
```

```python
def uniform_to_expo(uni_dist: np.ndarray, lmbda: float):
    return -np.log(uni_dist) / lmbda

# Sample exponential distriburion from uniform one 
u = np.random.uniform(0, 1, 1000)
expo = uniform_to_expo(u, 0.2)
```

```python
fig1 = plt.figure(figsize=(10, 10))

ax1 = fig1.add_subplot(2, 1, 1)
ax1.hist(u, color='red')
plt.title('Uniform Distribution')
plt.xlabel('u')
plt.ylabel('Freq.')

ax2 = fig1.add_subplot(2, 1, 2)
ax2.hist(expo, color='green')
plt.title('Exponential Distribution')
plt.xlabel('e')
plt.ylabel('Freq.')

plt.tight_layout()
```
